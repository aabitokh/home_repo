### Видно, что очень много звонков в районе 0 секунд, данные видимо отображаются неправильно, так как мобильный операторы фиксируют разговор только после  3 секунд. Лучше всего заполнить 0 медианой по абонентам. 

```python
def fill_calls_median(data):
    for type in calls['user_id'].unique():
        med = data.loc[data['user_id']==type]['duration'].median()
        data = data.replace(0, med)
    return data

"""функция подставляет вместо 0 медианнное 
значение по длительности разговоров для каждого клиента"""
```

### Получился датафрейм со всеми нужными данными и без NaNов, но его можно сделать более читаемым визуально. 

```python
final_data_with_tariffs = final_data_with_tariffs.pivot_table(index = [
    'user_id',
    'tariff',
    'city',
    'month'
], values = [
    'duration_sum',
    'messages', 
    'mb_used',
    'messages_included',
    'mb_per_month_included', 
    'minutes_included', 
    'rub_monthly_fee', 
    'rub_per_gb', 
    'rub_per_message',
    'rub_per_minute'
    
])
final_data_with_tariffs
```

```python
final_data_with_tariffs = final_data_with_tariffs.reset_index()
```

```python
final_data_with_tariffs
```

### Теперь посчитаем выручку с помощью функции ниже: 

```python
def revenue(row):
    if row['duration_sum']>row['minutes_included']:
        revenue_c = (row['duration_sum']-row['minutes_included'])*row['rub_per_minute']
    else:
        revenue_c = 0
    if row['messages']>row['messages_included']:
        revenue_m = (row['messages']-row['messages_included'])*row['rub_per_message']
    else:
        revenue_m = 0
    if row['mb_used']>row['mb_per_month_included']:
        gb = (row['mb_used']-row['mb_per_month_included'])/1024
        gb = ceil(gb)
        revenue_i = gb*row['rub_per_gb']
    else:
        revenue_i = 0
    revenue_all = revenue_m+revenue_i+revenue_c+row['rub_monthly_fee']
    return revenue_all
    
        
```

```python
final_data_with_tariffs['revenue'] = final_data_with_tariffs.apply(
    revenue, 
    axis = 1)
```

```python
final_data_with_tariffs
```

### Проверим различаются ли выручка по пользователям по тарифам. 

Hо: Выручки не отличаются друг от друга
H1: Выручки  отличаются друг от друга

```python
smart_revenue = smart_users['revenue'].mean()# введите ваш код тут

alpha = 0.05# введите ваш код тут

results = st.ttest_1samp(ultra_users['revenue'], smart_revenue)

print('p-значение: ', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Hо: выручки равны 

H1: выручка по тарифу ультра больше

```python
smart_revenue = smart_users['revenue'].mean()

alpha = 0.05

results = st.ttest_1samp(ultra_users['revenue'], smart_revenue)

print('p-значение:', results.pvalue/2)

if (results.pvalue/2 < alpha) and (ultra_users['revenue'].mean() > smart_revenue):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Скорее всего выручки отличаются, при этом ультра приносит больший доход компании, чем тариф смарт, возможно компании стоит сосредоточиться на продаже именно этого тарифа, подходящая база - большая клиентская база, которым тесно в тарифе смарт, например.

На первый взгляд выручки по регионам не сильно отличаются от московских, хоть последние и немноо больше говорят по телефону и пользуются интернетом, проверим является ли такое отличие по выручке - статистически значимым. 

Hо: Выручки не отличаются друг от друга

H1: Выручки  отличаются друг от друга

```python
msc_revenue = msc_group['revenue'].mean()# введите ваш код тут

alpha = 0.05# введите ваш код тут

results = st.ttest_1samp(region_users['revenue'], msc_revenue)

print('p-значение: ', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Опровергнуть нулевую гипотезу не получилось, логично - цены на тарифы одинаковые, россияне - везде россияне, пользуются +/- одинаково услугами, выручки сопоставимы.
